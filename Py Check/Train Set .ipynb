```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
df = pd.read_csv(r"C:\Users\arind\OneDrive\Desktop\Classes\T1 ESSEC Marketing Analytics\train.csv")
df.drop(['act_date','Unnamed: 0', 'calibration'],axis = 1,inplace=True)
df.set_index('contact_id',inplace=True) 
df.fillna('0', inplace=True)
df.head()
```

```python
X = df.drop('donation', axis = 1)
y = df['donation']
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.05, random_state=42)
```

```python

# Fitting Random Forest Regression to the dataset
# import the regressor
from sklearn.ensemble import RandomForestRegressor
  
 # create regressor object
regressor = RandomForestRegressor()
  
# fit the regressor with x and y data
regressor.fit(X_train, y_train)  
```

```python
pred = regressor.predict(X_test)
```

```python
li = []
for x in pred:
    if x<.5:
        li.append(0)
    else : li.append(1)
```

```python
print(X.shape)
len(li)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, li))

from sklearn.metrics import accuracy_score
accuracy_score(y_test, li)
```
